# Step 0: Environment Contract Check

This notebook verifies that `google-tunix`, `jax`, and `flax` are installed and detects the available hardware (TPU/GPU/CPU).

```python
import sys
import os
import jax

print(f"Python version: {sys.version}")

try:
    import tunix
    print(f"Tunix version: {tunix.__version__}")
except ImportError:
    print("Tunix not installed or not found in path.")

try:
    import flax
    print(f"Flax version: {flax.__version__}")
except ImportError:
    print("Flax not installed.")

print(f"JAX version: {jax.__version__}")
print(f"JAX devices: {jax.devices()}")

# Check for TPU
if any('TPU' in str(d) for d in jax.devices()):
    print("\u2705 TPU Detected!")
else:
    print("\u26a0\ufe0f No TPU detected. Running on CPU/GPU.")
```
